# REGRESSIONS

```python
df._get_numeric_data().columns.shape[0]
```

```python
features = [col for col in df._get_numeric_data().columns if col != 'SalePrice']
print('Number of features:',len(features))
```

```python
X = df[features]
y = df['SalePrice']
```

```python
print(X.shape)
print(y.shape)
```

```python
# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
# 75% training / 25% testing
```

## Scale Data

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

# BASELINE regressions (Linear, Ridge, Lasso)

```python
# LINEAR REG - Instantiate and score using cross validation (10 folds)
linreg = LinearRegression()
print('LINEAR REG Scores (BASELINE):')
print('X-Val score MEAN using X_train\t\t',cross_val_score(linreg, X_train, y_train,cv=10).mean())
print('X-Val score STD.DEV using X_train\t',cross_val_score(linreg, X_train, y_train,cv=10).std())
```

```python
# RIDGE - Instantiate and score using cross validation (10 folds)
ridge=RidgeCV(alphas=np.linspace(.1, 10, 100))
print('RIDGE Scores (BASELINE):')
print('X-Val score MEAN using X_train\t\t',cross_val_score(ridge, X_train, y_train,cv=10).mean())
print('X-Val score STD.DEV using X_train\t',cross_val_score(ridge, X_train, y_train,cv=10).std())
```

```python
# LASSO - Instantiate and score using cross validation (10 folds)
lasso = LassoCV(n_alphas=200,cv=3)
print('LASSO X-VAL Scores (BASELINE):')
print('X-Val score MEAN using X_train\t\t',cross_val_score(lasso, X_train, y_train,cv=10).mean())
print('X-Val score STD.DEV using X_train\t',cross_val_score(lasso, X_train, y_train,cv=10).std())
```

### Comments from BASELINE results:
- Linear regression fared the worst. Lasso has the highest score.
- Next step is to use Lasso to check the relative strengths of the coefficients.
- To pick the strongest coefficients for the next iteration of regression.

```python
# LASSO - Fit the model using training data
lasso.fit(X_train, y_train);
# this has now derived the optimum alpha and generated 177 coefficients
```

```python
# LASSO - Model scores (training data & test data)
print('LASSO Model Scores (BASELINE):')
print('Model score using X_train\t\t',lasso.score(X_train, y_train))
print('Model score using X_test\t\t',lasso.score(X_test, y_test))
```

```python
pd.Series(lasso.coef_, index=features).plot.bar(figsize=(15, 7));
```

```python
# create dataframe to store the coefficients
df_coef = pd.DataFrame(lasso.coef_,columns=['coeff'])
df_coef['feature_index'] = df_coef.index.values
# convert all coefficients to their absolute values
df_coef['coeff'] = [np.abs(i) for i in df_coef['coeff']]
```

```python
# plot labels above are too small to read... will enumerate & sort coefs to pick the features
df_coef = pd.DataFrame(lasso.coef_,columns=['coeff'])

# create an feature index column
df_coef['feature_index'] = df_coef.index.values

# change coefficient values to their absolute values
df_coef['coeff'] = [np.abs(i) for i in df_coef['coeff']]

# get the indexes of all coefficients that are > 0
feature_indexes = df_coef[df_coef['coeff']>0]['feature_index']
# feature_indexes is a Series object

print('Number of non-zero features\t',len(feature_indexes))
```

### Previously, we also concluded from the scatter plots that the following variables MAY have a higher order (squared) relationship with Sale Price
- Overall Qual
- Electrical
- 2nd Flr SF

Therefore, we will add a higher order (square) of these variables as our predictors to see if this improves the model. Using PolynomialFeatures would explode the number of predictors (which we want to avoid) due to the creation of interaction terms. So intead, we will add just the 3 predictors:

- 'Overall Qual Squared' = Overall Qual ^ 2
- 'Electrical Squared' = Electrical ^ 2
- '2nd Flr SF Squared' = 2nd Flr SF ^ 2

```python
df['Overall Qual Squared'] = df['Overall Qual']**2
df['Electrical Squared'] = df['Electrical']**2
df['2nd Flr SF Squared'] = df['2nd Flr SF']**2
```

# Prepare for Prediction

## Load TEST data to check that all features (derived from TRAINING data) are present

```python
df_test = pd.read_csv('../datasets/test_ss_r1.csv')
```

```python
# global variable
features_not_in_test_data = []
```

```python
# define function to handle KeyError exception that may arise in case columns in test data do not match training data
def key_error_handler(features_list, columns_list):
    # enable access to global variable
    global features_not_in_test_data
    # for each item in features_list
    for i in features_list:
        # check if the item is NOT in columns_list
        if i not in columns_list:
            # if so, remove item from features_list
            features_list.remove(i)
            # append the item into the global variable
            features_not_in_test_data.append(i)
    return features_list
```

```python
# This section checks if any of the selected features are missing from the test data
# This may occur for Nominal (unordered categorical) data, since dummy columns are 
# generated for each value of the data. If the test data does not have a particular value
# for this category, then the dummy column would be absent. As a result, we would not be able
# to index (filter) the DataFrame using that column name.

exception_resolved = False
exception_count = 0
df_columns_list = list(df_test.columns.values)
print(f'Starting with {len(features)} features')

while not exception_resolved:
    try:
        X = df_test[features]
    except KeyError:
        exception_count+=1
        # announce each instance that a KeyError exception is raised
        print(f'Resolving KeyError exception {exception_count}')
        features = key_error_handler(features,df_columns_list)
    else:
        # flag as resolved
        exception_resolved = True

# notify if new features were found in the test data
if len(features_not_in_test_data)>0:
    print(f'The following feature(s) is(are) absent from the test data:{features_not_in_test_data}')
    print(f'Proceed to train model without these features.')

# print the conclusion of this check
print(f'Next step: train model using {len(features)} features.')
```

## Train the model with the full set of training data & updated features

```python
len(features)
```

```python
df = pd.read_csv('../datasets/train_r1.csv')
df['Overall Qual Squared'] = df['Overall Qual']**2

# load training data (full set) with updated features set.
X = df[features]
y = df['SalePrice']
print(f'Full training predictors: {X.shape}')
print(f'Full training targets: {y.shape}')

# scale training data
ss = StandardScaler()
ss.fit(X)
X = ss.transform(X)

# RIDGE - Fit (train) the model using the full set of training data
ridge=RidgeCV(alphas=np.linspace(.1, 10, 100))
ridge.fit(X,y)

# this has now derived the optimum alpha and generated coefficients
print(f'coefficients:\n{ridge.coef_}')
print(f'optimum alpha:\n{ridge.alpha_}')
print(f'score:\n{ridge.score(X,y)}')
```
